### mbox-short (Part-1)

```python
path='mbox-short.txt'
file = open(path)

line=[]
line_idx=[]
count=0
rl=file.readlines()
for ln in range(len(rl)):
    count=count+1
    line_idx.append(len(rl[ln]))     # Adding length of each line into list
    line.append(count)               # Adding line numbers into list
print('\nTotal Lines:',count)        # Total No. of the lines
# # Trials
# print(line)                       # 'mbox-short.txt' file lines
# print('\n\n')
# print(len(line))                  # Total line numbers
# print('\n\n')
# print(line_idx)                   # Length of the each line.

# Testing
# for i in range(4):
#     print(rl[i],'\r')             # Test printing
    
print('\nLine - No', '   Max Length\n=========    ==========') 
for ln in range(len(line)):
    print(f'  {ln+1}      \t{line_idx[ln]}') 
```
