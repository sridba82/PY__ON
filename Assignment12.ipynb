- 'isalpha': only alphabets are True,  Numeric error, if convert numeric to apha then False
- 'isascii': all characters in text format is True, numeric error
- 'isdecimal': Only +ve interger values in text format is True, Numeric error
- 'isidentifier': alpha, starting alpha and numeric are True, Numeric error, text Numeric error
- 'islower': lower alpha, lower alpha and numeric in text are True, Numeric and Capital alpha error 
- 'isnumeric': only text numeric is True, Numeric error, alpha and alphanumeric False
- 'isprintable': except \n, \t all alpha, numeric, special char, symbols are True
- 'isspace': if single space in quotes is True, all are false
- 'istitle': only starting char capital in a string is True, rese all False. 
- 'isupper': If all alpabets are in capital then true, else false.
