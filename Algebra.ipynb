```python
import matplotlib.pyplot as plt
import time
import math
formula = plt.imread('C:\\Users\\Srinivas Rao\\a1.jpg')
plt.imshow(formula)
plt.axis('off')
plt.title('Formula') 
plt.show()
time.sleep(2)
a = eval(input("Enter 'a' value: "))
time.sleep(2)
b = eval(input("Enter 'b' value: "))
a_square = round(math.pow(a, 2))
b_square = round(math.pow(b, 2))
two_ab = 2*a*b
absquare= round(math.pow(a, 2)) + 2*a*b + round(math.pow(b, 2))
print(f'a square = {a_square}\nb square = {b_square}\n2ab = {two_ab}\nEquation Answer is: {absquare}')
```

```python
import matplotlib.pyplot as plt
import time
import math
formula = plt.imread('C:\\Users\\Srinivas Rao\\a3.jpg')
plt.imshow(formula)
plt.axis('off')
plt.title('Formula') 
plt.show()
time.sleep(2)
x = eval(input("Enter 'x' value: "))
x_sq = round(math.pow(x,2))
time.sleep(2)
y = eval(input("Enter 'y' value: "))
y_sq = round(math.pow(y,2))
x_sq_plus_y_sq = round((math.pow((x + y), 2) + math.pow((x-y),2)) / 2)
print(f'x square = {x_sq}\ny square = {y_sq}\nx Square + Y Square is:\n=======================\nFormula Answer is: {x_sq_plus_y_sq}')
```
