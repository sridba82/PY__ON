```python
# Q7-1 you need to read the text file: email.txt and separate fname, sname, company, email
path='emails.txt'
file=open(path)
lines=file.read()
emp_m_list=lines.split('; ')
emails=[]
for m in emp_m_list:
    emails.append(m[m.index('<')+1:m.index('>')])

f_name=[emails[i][:emails[i].index('.')] for i in range(len(emails))]
s_name=[emails[i][emails[i].index('.')+1 : emails[i].index('@')] for i in range(len(emails))]
c_name=[emails[i][emails[i].index('@')+1:][:emails[i][emails[i].index('@')+1:].index('.')] for i in range(len(emails))]
f_name, s_name, c_name, emails
```

```python
# Q7-2 you need to read the text file: email.txt and separate fname, sname, company, email
path='emails.txt'
file=open(path)
lines=file.read()
emp_m_list=lines.split('; ')
emails, f_name, l_name, company =[], [], [], []
for m in emp_m_list:
    emails.append(m[m.index('<')+1:m.index('>')])
for nsc in emails:
    company.append(nsc[nsc.index('@')+1:])
    f_name.append(nsc[:nsc.index('.')])
    l_name.append(nsc[nsc.index('.')+1:nsc.index('@')])


print('Firstname','\t','Surname', '\t\t','Company','\t\t','E-Mail')
print('===========','\t','===============', '   ','=============',' ','===================================')
for fetch in range(len(emails)):
    print(f_name[fetch].ljust(7,' '), '\t',l_name[fetch].ljust(17,' '), ' ', company[fetch].ljust(10,' '), ' ',emails[fetch])
```
